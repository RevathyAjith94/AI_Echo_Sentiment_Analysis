```python
import pandas as pd

df = pd.read_excel(r"D:\AI_Echo\data\raw\chatgpt_style_reviews_dataset.xlsx")
```

```python
df.dropna(subset=["review"], inplace=True)
```

Since review text is critical for NLP, rows with missing reviews were removed.

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words("english"))

def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^a-z\s]", "", text)
    tokens = text.split()
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return " ".join(tokens)

df["clean_review"] = df["review"].apply(clean_text)
```

Lemmatization was applied instead of stemming to preserve semantic meaning.
